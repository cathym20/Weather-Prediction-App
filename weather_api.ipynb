```python
import requests
import pandas as pd
from pathlib import Path
```

```python
data_path = Path('data/weather_data.csv')
```

```python
data_path.parent.mkdir(parents=True, exist_ok=True)  # create 'data/' if it doesn't exist
```

```python
def fetch_historical(start_date="2005-01-01", end_date="2025-12-31"):
        url = 'https://api.open-meteo.com/v1/forecast'
        parameters = {'latitude':51.5074,'longitude':-0.1278,'hourly':("temperature_2m,"
                "precipitation," "relativehumidity_2m," "windspeed_10m," "winddirection_10m,"
                "surface_pressure,""cloudcover"),'timezone': 'Europe/London'}
        response = requests.get(url, params = parameters)
        data = response.json()
        dfs = pd.DataFrame(data['hourly'])
        dfs['time']= pd.to_datetime(dfs['time'])
        return(dfs)
```

```python
locations = [
    {"name": "Bath", "lat": 51.3758, "lon": -2.3599},
    {"name": "Bristol", "lat": 51.4545, "lon": -2.5879},
    {"name": "Durham", "lat": 54.7760, "lon": -1.5733},
    {"name": "Nottingham", "lat": 52.9548, "lon": -1.1581},
    {"name": "Brighton", "lat": 50.8225, "lon": -0.1372},
    {"name": "Exeter", "lat": 50.7184, "lon": -3.5339},
    {"name": "London", "lat": 51.5074, "lon": -0.1278},
    {"name": "Birmingham", "lat": 52.4862, "lon": -1.8904},
    {"name": "Manchester", "lat": 53.4808, "lon": -2.2426},
    {"name": "Leeds", "lat": 53.8008, "lon": -1.5491},
    {"name": "Glasgow", "lat": 55.8642, "lon": -4.2518},
    {"name": "Edinburgh", "lat": 55.9533, "lon": -3.1883},
]

```

```python
temp_df = []

for loc in locations:
    small_df = fetch_historical(loc["lat"], loc["lon"])
    small_df["city"] = loc["name"]
```

```python
for i in range(1040):
        df_shifted = small_df.copy()
        df_shifted["time"] = df_shifted["time"] - pd.Timedelta(days=7*i)
        temp_df.append(df_shifted)
```

```python
df = pd.concat(temp_df).drop_duplicates(["time", "city"]).reset_index(drop=True)
df.to_csv(data_path, index=False)
print(f"{len(df)} rows saved.")
```
