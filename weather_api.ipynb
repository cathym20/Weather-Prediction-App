```python
import requests
import pandas as pd
from pathlib import Path
```

```python
data_path = Path('data/weather_data.csv')
```

```python
def fetch_historical(start_date="2023-01-01", end_date="2023-12-31"):
        url = 'https://api.open-meteo.com/v1/forecast'
        parameters = {'latitude':51.5074,'longitude':-0.1278,'hourly':("temperature_2m,"
                "precipitation," "relativehumidity_2m," "windspeed_10m," "winddirection_10m,"
                "surface_pressure,""cloudcover"),'timezone': 'Europe/London'}
        response = requests.get(url, params = parameters)
        data = response.json()
        df = pd.DataFrame(data['hourly'])
        df['time']= pd.to_datetime(df['time'])
        return(df)
```
